```python
#using if to define price ranges
def price_label(row):
    if row['prices'] <=5:
        return 'Low-range products'
    elif (row['prices'] >5) and (row['prices'] <=15):
        return 'Mid-range products'
    elif row['prices'] >15:
        return 'High range products'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#2 question: changing loop to add extra column for 2 busiest days and 2 slowest days
result = []

for value in df_ords_prods_merged['Order_day_of_week']:
    if value == 0:
        result.append('Two busiest days')
    elif value == 1:
        result.append('Two busiest days')
    elif value == 3:
        result.append('Two slowest days')
    elif value ==4:
        result.append('Two slowest days')
    else:
        result.append('Regular busy')
```

```python
result = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in range (9,17):
        result.append('Most orders')
    elif value in range(3,5):
        result.append('Fewest orders')
    else:
        result.append('Average orders')
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result
```
