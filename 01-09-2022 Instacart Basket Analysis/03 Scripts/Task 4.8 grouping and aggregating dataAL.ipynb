```python
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'order_busy_frequencies.pkl'))
```

```python
df_ords_prods_merge = df_ords_prods_merge.drop(df_ords_prods_merge.columns[[2]], axis=1)
```

```python
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

The subset dataframe gave us means from 8 of 21 departments that ranged between 11-19. Now running the mean for the full report, since we have a larger order count the means of all 21 departments have now risen between 15-22

```python
#4 question: create a loyalty flag
df_ords_prods_merge['max order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max order'] >40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max order'] <=40) & (df_ords_prods_merge['max order'] >10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max order'] <=10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag']. value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

```python
#5 question: product price statistics about all 3 loyalty categories
```

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min']})
```

```python
#6 question: : establishing the mean of prices and developping a user spending category
df_ords_prods_merge['Mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['Mean_price'] <10, 'Spend_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['Mean_price'] >=10, 'Spend_flag'] = 'High spender'
```

```python
df_ords_prods_merge['Spend_flag']. value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

```python
#7 question: establishing the median of the "Days since last order' and create of order frequency flag
df_ords_prods_merge['Median_since_last_ordered'] = df_ords_prods_merge.groupby(['user_id'])['Days since last order'].transform(np.median)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['Median_since_last_ordered'] >20, 'Order_frequency_flag'] = 'Non-frequent cust'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['Median_since_last_ordered']>10) & (df_ords_prods_merge['Median_since_last_ordered']<=20), 'Order_frequency_flag'] = 'Regular cust'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['Median_since_last_ordered'] <=10, 'Order_frequency_flag'] = 'Frequent cust'
```

```python
df_ords_prods_merge['Order_frequency_flag']. value_counts(dropna = False)
```
